## Simple Plot

```python
import matplotlib.pyplot as plt
```

```python
# plot
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.title('Title')
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]
plt.plot(x, y)
```

```python
import numpy as np

x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)
```

## Implicit

```python
# pyplot style
plt.subplot(1,2,1) #nrow, ncol, index
plt.plot(x,y1)
plt.title('Sine')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Cosine')
```

## Explicit

```python
# object-oriented (OO) style

# Create a figure
fig = plt.figure()

# Define the positions and sizes of the axes within the figure
# [left, bottom, width, height]
axes1_position = [0.2, 0.1, 0.8, 0.4]
axes2_position = [0.1, 0.6, 0.4, 0.3]

axes1 = fig.add_axes(axes1_position)
axes2 = fig.add_axes(axes2_position)

# Plot data in each Axes
axes1.plot(x, y1)
axes1.set_title('Sine')
axes1.set_xlabel('x-axis')
axes1.set_ylabel('y-axis')

axes2.plot(x, y2)
axes2.set_title('Cosine')
axes2.set_xlabel('x-axis')
axes2.set_ylabel('y-axis')
```

```python
# Create a 2x2 grid of subplots
fig, ax = plt.subplots(2, 2) # row, col

# Plot data in each subplot
ax[0, 0].plot(x, y1)
ax[0, 0].set_title('Sine')
ax[0, 1].plot(x, y2)
ax[0, 1].set_title('Cosine')
ax[1, 0].plot(x, y3)
ax[1, 0].set_title('Tangent')
ax[1, 1].plot(x, y4)
ax[1, 1].set_title('Exponential')

# Adjust layout and display the plot
plt.tight_layout()
```

## artist style

```python
data1, data2, data3, data4 = np.random.randn(4, 100)  # make 4 random data sets

fig, ax = plt.subplots(figsize=(5, 2.7)) # width, height(inch)
x = np.arange(len(data1))

ax.plot(x, np.cumsum(data1), color='blue', linewidth=3, linestyle='--')
l, = ax.plot(x, np.cumsum(data2), color='orange', linewidth=2)
l.set_linestyle(':')
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7)) # width, height(inch)
ax.plot(x, np.cumsum(data1),
        color='blue',
        linewidth=3,
        linestyle='--',
        marker="+",
        label="ax1")
l, = ax.plot(x, np.cumsum(data2), color='orange', linewidth=2, label='ax2')
l.set_linestyle(':')
ax.legend()
```
