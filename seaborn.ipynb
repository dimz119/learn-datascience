```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# examples are available on https://github.com/mwaskom/seaborn-data
data = sns.load_dataset('titanic')
```

```python
data.head()
```

```python
# distribution
sns.histplot(data=data, x='age')
```

```python
sns.kdeplot(data=data, x='age')
```

```python
sns.rugplot(data['age'])
```

```python
sns.scatterplot(data=data, x='age', y='fare')
plt.show()
```

```python
# Categorial value
sns.barplot(data=data, x='class', y='fare')
plt.show()
```

```python
# categorial value
sns.boxplot(data=data, x='class', y='age', hue="survived")
plt.title('Class vs Age in Titanic Passengers')
plt.xlabel('Class')
plt.ylabel('Age')
sns.set_style('whitegrid')
plt.show()
```

```python
sns.violinplot(data=data, x='class', y='age', hue="survived")
```

```python
sns.jointplot(data=data, x='age', y='fare', kind='scatter')
```

```python
sns.pairplot(data=data.dropna(), vars=['age', 'fare'], hue='survived')
plt.show()
```

```python
correlation_matrix = data.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap for Titanic Dataset')
plt.show()
```

```python
g = sns.FacetGrid(data=data, col='pclass', row='survived', hue='sex', margin_titles=True)
g.map(sns.histplot, 'age', bins=20, alpha=0.6)
g.add_legend()
plt.show()
```
