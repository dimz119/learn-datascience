```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Generate a sample dataset with 300 samples, 2 features, and 4 clusters
data, labels = make_blobs(
    n_samples=300,
    centers=4,
    n_features=2,
    cluster_std=1.9,
    random_state=42)
# n_samples (int, optional, default=100): The total number of points to generate, divided among the clusters.
# n_features (int, optional, default=2): The number of features (dimensions) for each data point.
# centers (int or array-like, optional, default=None): The number of centers (clusters) to generate, or the fixed center locations. If None, 3 centers are generated.
# cluster_std (float or sequence of floats, optional, default=1.0): The standard deviation of each cluster. A single float value will be applied to all clusters, while a sequence of floats will be applied to each cluster individually.
```

```python
# Specify the number of clusters (K)
num_clusters = 4

# Create a KMeans instance with the specified number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
```

```python
# Fit the KMeans model to the dataset
kmeans.fit(data)
```

```python
# Obtain the coordinates of the cluster centroids
centroids = kmeans.cluster_centers_
print("Cluster centroids:", centroids)
```

```python
# Obtain the cluster assignments for each data point
assignments = kmeans.labels_
# Print the cluster assignments and centroids
print("Cluster assignments:", assignments)
```
